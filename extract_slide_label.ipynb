extract the images containing the slide name

```python
import os
import glob
from openslide import OpenSlide
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
case = "AdCC"
wsi_folder_path = os.path.join("/home/john/Data/WSI/Sheffield cases", case)

files = []
for file_path in sorted(glob.glob(f"{wsi_folder_path}/*svs")):
    print(file_path)
    files.append(file_path)
```

```python
for f in files:
    try:
        o = OpenSlide(f)
        img = o.associated_images
    except:
        continue
    if not img["label"]:
        print(f"Error finding label for {f}")
        continue
    label = img["label"]
    label = label.rotate(-90)
    img_array = np.array(label)
    plt.imshow(img_array)
    plt.axis('off')

    # save it
    filename = os.path.basename(f)
    print(filename)
    plt.savefig(os.path.join("labels", f"{case}_{filename}.png"), bbox_inches='tight', pad_inches = 0)

    # break
```
